```python

from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
```

```python
hdulist = fits.open('140311A.fits')
```

```python
Flux = hdulist[0]
Error = hdulist[1]
Wavelength = hdulist[2]
```

```python
plt.figure()
plt.errorbar(Wavelength.data, Flux.data, yerr= Error.data, ecolor='r')
```

```python
plt.show()
```
